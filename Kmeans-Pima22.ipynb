```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
%matplotlib inline
import pandas as pd
cust_df = pd.read_csv("pima-indians-diabetes.csv", names=['Pregnancies','Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Class'])
cust_df.head(10)
```

```python
df = cust_df.drop('Class', axis=1)
df.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
X = df.values[:,:]
X = np.nan_to_num(X)
Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet
```

```python
clusterNum = 2
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
df["Clus_km"] = labels
df.head(10)
```

```python
df.groupby('Clus_km').mean()
```

```python
area = np.pi * ( X[:, 1])**2  
plt.scatter(X[:, 1], X[:, 4], c=labels.astype(np.float), alpha=0.5)
plt.xlabel('Glucose', fontsize=18)
plt.ylabel('Insulin', fontsize=16)

plt.show()
```

```python
from sklearn.metrics import silhouette_score
print(f'Silhouette Score(n=2): {silhouette_score(X, labels)}')
```

```python
silhouette_data = []
for k in range (2,13):
    kmeans_shihouette = KMeans(init = "k-means++", n_clusters = k, n_init = 12)
    kmeans_shihouette.fit(X)
    shihouette_labels = kmeans_shihouette.labels_
    silhouette_data.append(silhouette_score(X, shihouette_labels))
    
plt.plot(range(2,13), silhouette_data, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Silhouette Score Pima')
plt.show
```
